```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('winequality.csv')
for col, value in df.items():
    if col != 'type':
        df[col] = df[col].fillna(df[col].mean())
df_blancos = df[df['type'] == 'white']
# Guarda los vinos blancos en un nuevo archivo CSV
df_blancos.to_csv('vinos_blancos.csv', index=False)  
df_blancos['quality'].unique()
```

```python
df_blancos['quality'] = [1 if x > 7 else 0 for x in df_blancos['quality']]
df_blancos['quality'].unique()
```

```python
X =df_blancos.drop(['type', 'quality'], axis=1)
y = df_blancos['quality']
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import plot_model

# Equilibrio de clases
oversample = SMOTE(k_neighbors=4)
X, y = oversample.fit_resample(X, y)

# División del conjunto de datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Escalamiento de características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Definir la arquitectura de la red neuronal
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(11,)))
model.add(Dropout(0.2))  # Capa de dropout para regularización
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compilar el modelo
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)  # Ajuste de la tasa de aprendizaje
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
model.fit(X_train, y_train, epochs=25, batch_size=32, validation_data=(X_test, y_test))

# Evaluar el modelo
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Loss: {loss}')
print(f'Accuracy: {accuracy}')

# Print the model summary
model.summary()

#Save model
tf.saved_model.save(model, 'saved_model')
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import joblib
from tkinter import *
from tensorflow import keras

# Carga del modelo
model = joblib.load('wine_quality_prediction_model_blanco')

# Función para realizar la predicción
def predict_quality():
    # Obtener los valores de los campos de entrada
    fixed_acidity = float(e1.get())
    volatile_acidity = float(e2.get())
    citric_acid = float(e3.get())
    residual_sugar = float(e4.get())
    chlorides = float(e5.get())
    free_sulfur_dioxide = float(e6.get())
    total_sulfur_dioxide = float(e7.get())
    density = float(e8.get())
    pH = float(e9.get())
    sulphates = float(e10.get())
    alcohol = float(e11.get())

    # Escalar las características de entrada
    scaler = StandardScaler()
    input_data = np.array([[fixed_acidity, volatile_acidity, citric_acid, residual_sugar, chlorides,
                            free_sulfur_dioxide, total_sulfur_dioxide, density, pH, sulphates, alcohol]])
    input_data = scaler.fit_transform(input_data)

    # Realizar la predicción
    prediction = model.predict(input_data)

    if prediction[0] == 0:
        result_label.config(text="Bad Quality Wine")
    else:
        result_label.config(text="Good Quality Wine")

# Configuración del GUI
master = Tk()
master.title("Wine Quality Prediction")

# Etiquetas y campos de entrada
labels = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides",
          "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol"]
entries = []

for i, label in enumerate(labels):
    Label(master, text=label).grid(row=i, column=0)
    entry = Entry(master)
    entry.grid(row=i, column=1)
    entries.append(entry)

# Botón para realizar la predicción
predict_button = Button(master, text="Predict", command=predict_quality)
predict_button.grid(row=len(labels), columnspan=2, pady=10)

# Etiqueta para mostrar el resultado de la predicción
result_label = Label(master, text="", font=("Arial", 16), pady=20)
result_label.grid(row=len(labels)+1, columnspan=2)

# Ejecutar el bucle principal del GUI
mainloop()
```
